```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bb = pd.read_csv(f, sep='\t', names=['#CHR', 'START', 'END', 'SAMPLE', 'RD', '#SNPS', 'COV', 'ALPHA', 'BETA', 'BAF', 'CLUSTER', "cn_normal", "u_normal", "cn_clone1", "u_clone1", "cn_clone2", "u_clone2"], skiprows=1)
bb['#CHR'] = bb['#CHR'].apply(lambda x: int(x[3:]))
```

```python
bb1 = bb.copy()
bb2 = bb.copy()
```

```python
bb1['sum_cn'] = bb['cn_clone1'].apply(lambda x: sum(np.array(x.split('|')).astype(int)) - 0.15)
bb1['which_cn'] = bb['cn_clone1'].apply(lambda x: "Clone 1")

bb2['sum_cn'] = bb['cn_clone2'].apply(lambda x: sum(np.array(x.split('|')).astype(int)))
bb2['which_cn'] = bb['cn_clone2'].apply(lambda x: "Clone 2")
```

```python
bb1
```

```python
bb2
```

```python
bb3 = pd.concat([bb1, bb2])
bb3
```

```python
bb_proc = bb3[bb3['SAMPLE'] == 'P10_DCIS']

# Get the start of each chromosome in order to calculate the genome position of each bin
chr_starts = {}
chrs = list(bb["#CHR"].unique()) # 1 - 22
chrs.sort()
length = 0
for chrm in chrs:
    chr_starts[chrm] = length
    length += bb[bb['#CHR'] == chrm]['END'].max()
bb_proc['genomePosition'] = bb.apply(lambda row: row['START'] + chr_starts[row['#CHR']], axis = 1)
```

```python
bb_proc
```

```python
chr_starts
```

```python
sns.lmplot(data=bb_proc, x='genomePosition', y='sum_cn', fit_reg=False, height=2, aspect=4, hue="which_cn", palette=["#E66100", "#5D3A9B"], scatter_kws={"s": 1}, legend=False)

plt.gca().set_ylabel("Read Depth Ratio")
plt.gca().set_xlabel("Chromosome")
plt.gcf().tight_layout()
# plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
# plt.legend(loc="lower center", bbox_to_anchor=(0.5, -1)) #, borderaxespad=0.)

maxY = bb_proc['RD'].max()

odd_chr_starts = dict(filter(lambda chr_start: chr_start[0] % 2 != 0, chr_starts.items())) 
even_chr_starts = dict(filter(lambda chr_start: chr_start[0] % 2 == 0, chr_starts.items())) 
y_ticks = [0, 1, 2, 3, 4]
y_labels = [0, 1, 2, 3, 4]
# x_ticks = list(odd_chr_starts.values())
x_ticks = list(chr_starts.values())
plt.gca().set_xticks(x_ticks)
# plt.gca().set_xticklabels(list(odd_chr_starts.keys()))
plt.gca().set_xticklabels(list(chr_starts.keys()))
plt.gca().set_yticks(y_ticks)
plt.gca().set_yticklabels(y_labels)
plt.gcf().set_size_inches((14.5,3))

chr_lines = list(chr_starts.values())
plt.gca().vlines(x=chr_lines, ymin=-1, ymax=maxY + 2, linestyles='dashed', lw=0.5)
plt.gca().set_ylim((0, 4.5))
plt.savefig('/Users/gillianchu/mek/cnaviz/genes/realfig/p10_dcis_hatchet_rdr.svg', bbox_inches='tight', dpi=300)
```
